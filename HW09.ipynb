Використаємо два надані файли: dataset_2d.csv для двовимірного датасету і mnist.csv для датасету MNIST. Для обох випадків ми будемо використовувати алгоритм K-means і метод ліктя для визначення оптимальної кількості кластерів.

для 'data_2d.csv'

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Завантаження даних
dataset_2d_path = extracted_dir_path + 'data/' + 'data_2d.csv'
df_2d = pd.read_csv(dataset_2d_path)

# Підготовка даних (стандартизація)
scaler = StandardScaler()
scaled_data_2d = scaler.fit_transform(df_2d)

# Використання методу ліктя для визначення оптимальної кількості кластерів
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data_2d)
    sse.append(kmeans.inertia_)

# Графік методу ліктя
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Кількість кластерів')
plt.ylabel('SSE')
plt.title('Метод ліктя для визначення оптимальної кількості кластерів (для dataset_2d)')
plt.grid(True)
plt.show()
```

для 'mnist.csv'

```python
# Завантаження даних
mnist_path = extracted_dir_path + 'data/' + 'mnist.csv'
df_mnist = pd.read_csv(mnist_path)

# Підготовка даних (не потрібна стандартизація, оскільки MNIST вже масштабований)

# Використання методу ліктя для визначення оптимальної кількості кластерів
sse_mnist = []
for k in range(1, 11):
    kmeans_mnist = KMeans(n_clusters=k, random_state=42)
    kmeans_mnist.fit(df_mnist)
    sse_mnist.append(kmeans_mnist.inertia_)

# Графік методу ліктя
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse_mnist, marker='o')
plt.xlabel('Кількість кластерів')
plt.ylabel('SSE')
plt.title('Метод ліктя для визначення оптимальної кількості кластерів (для MNIST)')
plt.grid(True)
plt.show()
```

## Завдання 2

Візуалізуйте результат роботи кластеризації. Для випадку з mnist датасетом, вам потрібно ще скористатись алгоритмом PCA щоб зменшити розмірність вашим даних до 2-вимірного варіанту.

Для двовимірного датасету (data_2d.csv):

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Завантаження даних
dataset_2d_path = extracted_dir_path + 'data/' + 'data_2d.csv'
df_2d = pd.read_csv(dataset_2d_path)
```

```python
df_2d
```

```python
df_2d.columns = ['Feature 1', 'Feature 2', 'cluster']
```

```python
df_2d.head()
```

```python




# Підготовка даних (стандартизація)
scaler = StandardScaler()
scaled_data_2d = scaler.fit_transform(df_2d)

# Кластеризація з використанням оптимальної кількості кластерів (визначеної з методу ліктя)
k = 3  # наприклад, для прикладу виберемо кількість кластерів як 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(scaled_data_2d)
df_2d['cluster'] = kmeans.labels_

# Візуалізація результатів кластеризації
plt.figure(figsize=(10, 6))
plt.scatter(df_2d['Feature 1'], df_2d['Feature 2'], c=df_2d['cluster'], cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red', label='Центроїди')
plt.title('Кластеризація двовимірного датасету з використанням K-means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

Для датасету MNIST (mnist.csv) з використанням PCA:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Завантаження даних
mnist_path = extracted_dir_path + 'data/' + 'mnist.csv'
df_mnist = pd.read_csv(mnist_path)
```

```python
df_mnist.head()
```

```python
# Додавання підписів колонок
# Заголовки колонок для MNIST можуть бути у вигляді 'pixel0', 'pixel1', ..., 'pixel783'
pixel_columns = [f'pixel{i}' for i in range(784)]
df_mnist.columns = ['label'] + pixel_columns  # Додаємо 'label' як заголовок для цільової змінної

# Перевірка перших рядків для підтвердження
print(df_mnist.head())
```

```python
# Розділення на ознаки та цільову змінну
X = df_mnist.drop('label', axis=1)
y = df_mnist['label']
```

```python
# Стандартизація даних
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Використання PCA для зменшення розмірності до 2 вимірів
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Кластеризація з використанням оптимальної кількості кластерів (визначеної з методу ліктя)
k = 10  # наприклад, для прикладу виберемо кількість кластерів як 10
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)
df_mnist['cluster'] = kmeans.labels_

# Візуалізація результатів кластеризації
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_mnist['cluster'], cmap='viridis', edgecolor='k', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red', label='Центроїди')
plt.title('Кластеризація датасету MNIST з використанням K-means та PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
